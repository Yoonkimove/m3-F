# 디큐브냐 타임스퀘어냐

```python
comparison = df.loc[(df.Name == '타임스퀘어') | (df.Name =='디큐브시티')]
comparison.head()

# test['Name'].str.extract(r'^(스타필드|롯데몰|롯데백화점|롯데마트|이마트|신세계백화점|현대백화점)\s.*')
```

```python
plt.figure(figsize=(20,10))

# sns.set(font_scale=2)

# plt.rc('xtick', labelsize=15) 
# plt.rc('ytick', labelsize=15)

ax = sns.countplot(x='Name', hue ='Category', data=comparison, palette="Pastel1")

plt.setp(ax.get_legend().get_texts(), fontsize='16')
plt.setp(ax.get_legend().get_title(), fontsize='20')

ax.axes.set_title("\n타임스퀘어와 디큐브시티의 카테고리 비교\n",fontsize=28)
ax.set_xlabel("Name",fontsize=20)
ax.set_ylabel("Count",fontsize=20)
ax.tick_params(labelsize=15)

ax
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(x='Name', hue ='Franchise', data=comparison, palette="Set2")

plt.setp(ax.get_legend().get_texts(), fontsize='16')
plt.setp(ax.get_legend().get_title(), fontsize='20')

ax.axes.set_title("\n타임스퀘어와 디큐브시티의 프랜차이즈 비교\n",fontsize=28)
ax.set_xlabel("Name",fontsize=20)
ax.set_ylabel("Count",fontsize=20)
ax.tick_params(labelsize=15)

ax
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(x='Name', hue ='Luxury', data=comparison, palette="Set2")

plt.setp(ax.get_legend().get_texts(), fontsize='16')
plt.setp(ax.get_legend().get_title(), fontsize='20')

ax.axes.set_title("\n타임스퀘어와 디큐브시티의 명품매장 비교\n",fontsize=28)
ax.set_xlabel("Name",fontsize=20)
ax.set_ylabel("Count",fontsize=20)
ax.tick_params(labelsize=15)

ax
```
