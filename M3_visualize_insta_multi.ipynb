# <font color = 'orange'> folium을 이용해서 대형몰(multiprex) 데이터 지도시각화

우리는 서울인근에 있는 대형몰들을 한 눈에 보기 위해 지도에 찍어보기로 했다.  
그 중 10개를 선정하였고 folium 기법을 이용하여 지도시각화를 해보겠다.  

**사용한 folium기법**  
* choropleth --> 인스타그램태그량
* Marker --> 10대 대형몰 위치(같은 회사는 같은 색깔로 표시)
* Circle --> 백화점, 대형 마트(같은 회사는 같은 색깔로 표시)


```python
import json
import folium
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```python
import folium
geo_df = test

map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Mapbox Bright')

map.choropleth(geo_data=geo_str,
              data=guDat,
              columns=['Gu','counts'],
              fill_color='GnBu', #PuRd, YlGnBu
              key_on='feature.properties.name',
              highlight=True,
              line_color='olivedrab',
               line_weight=0.15,
               legend_name='인스타그램태그량'
              )

for i, row in geo_df.iterrows():
    mall_name = row['Name'] + '-' + row['Address']
    if row['Category'] == '롯데몰':
        icon_color='purple'      
    elif row['Category'] == '스타필드':
        icon_color='orange'
    else:
        icon_color='darkblue'
        
    
    icon = folium.Icon(color=icon_color,icon='info-sign')
        
    if row['Category'] == '롯데백화점':
        location = (row['Latitude'], row['Longitude']) 
        folium.Circle(
            location=location,
            radius=350,
            color='slateblue',
            weight=1,
            fill_opacity=0.6,
            opacity=1,
            fill_color='slateblue',
            popup=mall_name,
            fill=True
        ).add_to(map)
    elif row['Category'] == '신세계백화점':
        location = (row['Latitude'], row['Longitude']) 
        folium.Circle(
            location=location,
            radius=350,
            color='mediumseagreen',
            weight=1,
            fill_opacity=0.6,
            opacity=1,
            fill_color='mediumseagreen',
            popup=mall_name,
            fill=True
        ).add_to(map)
    elif row['Category'] == '이마트':
        location = (row['Latitude'], row['Longitude']) 
        folium.Circle(
            location=location,
            radius=350, 
            color='forestgreen',
            weight=1,
            fill_opacity=0.6,
            opacity=1,
            fill_color='forestgreen',
            popup=mall_name,
            fill=True
        ).add_to(map)
    elif row['Category'] == '현대백화점':
        location = (row['Latitude'], row['Longitude']) 
        folium.Circle(
            location=location,
            radius=350, 
            color='palevioletred',
            weight=1,
            fill_opacity=0.6,
            opacity=1,
            fill_color='palevioletred',
            popup=mall_name,
            fill=True
        ).add_to(map)
    elif row['Category'] == '롯데마트':
        location = (row['Latitude'], row['Longitude']) 
        folium.Circle(
            location=location,
            radius=350, # 지름이 degree에 비례하도록 설정
            color='chocolate',
            weight=1,
            fill_opacity=0.6,
            opacity=1,
            fill_color='chocolate',
            popup=mall_name,
            fill=True
        ).add_to(map)
    else:
        folium.Marker(
            [row['Latitude'], row['Longitude']], 
            popup=mall_name,
            icon=icon
        ).add_to(map)
    
map
```
